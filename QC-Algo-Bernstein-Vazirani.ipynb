```python
from qiskit import *
```

```python
from qiskit.visualization import plot_histogram
```

```python
%matplotlib inline
```

```python
secret = input()
```

```python
# temp value: extra 1 qubit
circuit = QuantumCircuit(len(secret) + 1, len(secret))
```

```python
circuit.draw(output="mpl")
```

```python
circuit.h(list(range(len(secret))))
circuit.draw(output="mpl")
```

```python
circuit.x(len(secret))
circuit.h(len(secret))
```

```python
circuit.barrier()
```

```python
#oracle
for index,i in enumerate(secret):
  if i == '1':
    circuit.cx(index, len(secret)) 
```

```python
circuit.barrier()
circuit.draw(output="mpl")
```

```python
circuit.h(list(range(len(secret))))
circuit.draw(output="mpl")
```

```python
circuit.measure(list(range(len(secret))), list(range(len(secret))))
circuit.draw(output="mpl")
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(circuit, simulator, shots=100).result()
```

```python
result.get_counts()
```
